```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix, roc_curve, auc


```

```python
X_2023 = df_2023.drop(['id','Class'], axis=1, errors = 'ignore')
y_2023 = df_2023['Class']
# drop the id column and split the data
```

```python
print(X_2023.shape)
print(y_2023.shape)
print(X_2023.columns.tolist())
```

```python
# draw random data to train and test
X_train_2023, X_test_2023, y_train_2023, y_test_2023 = train_test_split(X_2023, y_2023, random_state=42, test_size=0.2)
```

```python
#check trainig and testing data shape
X_train_2023.shape
```

```python
X_test_2023.shape
```

```python
scaler = StandardScaler()
X_train_2023 = scaler.fit_transform(X_train_2023)
X_test_2023 = scaler.transform(X_test_2023)
```

```python
print(pd.Series(y_train_2023).value_counts(normalize=True))
```

```python
rf_2023 = RandomForestClassifier(
    n_estimators=100,
    max_depth= 10,
    min_samples_split = 5,
    random_state=42,
    n_jobs=-1)
rf_2023.fit(X_train_2023, y_train_2023)
```

```python
cv_score_2023 = cross_val_score(rf_2023, X_train_2023, y_train_2023, cv=5, scoring='f1')
print("Cross validation F1 score:",cv_score_2023)
print("Average F1 score: ",cv_score_2023.mean())
```

```python
## prediction for second model
y_pred= rf_2023.predict(X_test_2023)
print(classification_report(y_test_2023, y_pred))
accuracy_score(y_test_2023, y_pred)
print(f"Accuracy for Second model: {accuracy_score(y_test_2023, y_pred)}")
```
